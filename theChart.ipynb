```python
print("Stock price keys:", list(prices_dict.keys()))
print("News headline keys:", list(headlines_dict.keys()))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

companies = ['AAPL', 'GOOG', 'NVDA', 'XOM', 'MCD']

for company in companies:
    df_price = prices_dict[company].copy()
    df_news = headlines_dict[company].copy()
    
    df_price['date'] = pd.to_datetime(df_price['date'])
    df_news['Date'] = pd.to_datetime(df_news['Date'])
    
    news_counts = df_news.groupby(df_news['Date'].dt.date).size()
    
    fig, ax1 = plt.subplots(figsize=(12,6))
    
    ax1.plot(df_price['date'], df_price['close'], color='blue', label='Close Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Close Price', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    
    ax2 = ax1.twinx()
    ax2.bar(news_counts.index, news_counts.values, color='orange', alpha=0.3, label='News Count')
    ax2.set_ylabel('Number of News Headlines', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    
    plt.title(f"{company}: Stock Price vs Number of News Headlines")
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    
    plt.show()
```

```python
keywords = ["increase", "profit", "positive", "upgrade"] 

for company in companies:
    df_price = prices_dict[company].copy()
    df_news = headlines_dict[company].copy()
    
    df_price['date'] = pd.to_datetime(df_price['date'])
    df_news['Date'] = pd.to_datetime(df_news['Date'])
    
    pattern = '|'.join(keywords)  
    df_news_filtered = df_news[df_news['Article_title'].str.contains(pattern, case=False, na=False)]
    
    news_counts = df_news_filtered.groupby(df_news_filtered['Date'].dt.date).size()
    
    fig, ax1 = plt.subplots(figsize=(12,6))
    
    ax1.plot(df_price['date'], df_price['close'], color='blue', label='Close Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Close Price', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    
    ax2 = ax1.twinx()
    ax2.bar(news_counts.index, news_counts.values, color='orange', alpha=0.3, label=f"News Count with keywords")
    ax2.set_ylabel('Number of News Headlines', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    
    plt.title(f"{company}: Stock Price vs News Headlines containing keywords")
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    
    plt.show()
```
